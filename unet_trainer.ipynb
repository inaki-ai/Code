```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use('extensys-gd')
```

```python
data = pd.read_csv('experiments/exp7/metrics.csv')
```

```python
# Train y val loss
FROM = 0
epochs = data['epoch']
train_loss = data['train_loss']
val_loss = data['val_loss']

fig, ax = plt.subplots(1, 1, figsize=(15,6))

ax.plot(epochs[FROM:], train_loss[FROM:], label='Train loss')
ax.plot(epochs[FROM:], val_loss[FROM:], label='Val loss')

ax.legend(loc='upper right')

#ax.set_yticks(np.arange(0, 1.1, 0.1))

ax.set_xlabel('epoch')
ax.set_ylabel('loss')

plt.show()
```

```python
# CCR
FROM = 0
epochs = data['epoch']
ccr = data['ccr']

fig, ax = plt.subplots(1, 1, figsize=(15,6))

ax.plot(epochs[FROM:], ccr[FROM:], label='CCR')

ax.legend(loc='lower right')

ax.set_xlabel('epoch')
ax.set_ylabel('CCR')

plt.show()
```

```python
# Precision recall
FROM = 0
epochs = data['epoch']
precision = data['precision']
recall = data['recall']

fig, ax = plt.subplots(1, 2, figsize=(15,6))

ax[0].plot(epochs[FROM:], precision[FROM:], label='Precision')
ax[1].plot(epochs[FROM:], recall[FROM:], label='Recall')

ax[0].legend(loc='lower right')
ax[1].legend(loc='lower right')

ax[0].set_xlabel('epoch')
ax[1].set_xlabel('epoch')

ax[0].set_ylabel('precision')
ax[1].set_ylabel('recall')

plt.show()
```

```python
# CCR
FROM = 0
epochs = data['epoch']
specifity = data['specifity']

fig, ax = plt.subplots(1, 1, figsize=(15,6))

ax.plot(epochs[FROM:], specifity[FROM:], label='specifity')

ax.legend(loc='lower right')

ax.set_xlabel('epoch')
ax.set_ylabel('specifity')

plt.show()
```

```python
# CCR
FROM = 0
epochs = data['epoch']
f1_score = data['f1_score']

fig, ax = plt.subplots(1, 1, figsize=(15,6))

ax.plot(epochs[FROM:], f1_score[FROM:], label='f1_score')

ax.legend(loc='lower right')

ax.set_xlabel('epoch')
ax.set_ylabel('f1_score')

plt.show()
```

```python
# Precision recall
FROM = 0
epochs = data['epoch']
jaccard_coef = data['jaccard_coef']
dsc_coef = data['dsc_coef']

fig, ax = plt.subplots(1, 2, figsize=(15,6))

ax[0].plot(epochs[FROM:], jaccard_coef[FROM:], label='jaccard_coef')
ax[1].plot(epochs[FROM:], dsc_coef[FROM:], label='dsc_coef')
ax[1].plot(epochs[FROM:], [0.8 for _ in epochs[FROM:]], '--')
ax[1].plot(epochs[FROM:], [0.85 for _ in epochs[FROM:]], '--')

ax[0].legend(loc='lower right')
ax[1].legend(loc='lower right')

ax[0].set_xlabel('epoch')
ax[1].set_xlabel('epoch')

ax[0].set_ylabel('jaccard_coef')
ax[1].set_ylabel('dsc_coef')

plt.show()
```

```python
# Precision recall
FROM = 0
epochs = data['epoch']
roc_auc = data['roc_auc']
pr_auc = data['pr_auc']

fig, ax = plt.subplots(1, 2, figsize=(15,6))

ax[0].plot(epochs[FROM:], roc_auc[FROM:], label='roc_auc')
ax[1].plot(epochs[FROM:], pr_auc[FROM:], label='pr_auc')

ax[0].legend(loc='lower right')
ax[1].legend(loc='lower right')

ax[0].set_xlabel('epoch')
ax[1].set_xlabel('epoch')

ax[0].set_ylabel('roc_auc')
ax[1].set_ylabel('pr_auc')

plt.show()
```
