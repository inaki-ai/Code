```python
imagen = Image.open('/home/imartinez/Dataset_TFM/images/DatasetB/benign_000005.png').convert("L")
imagen = Image.open('jirafa.jpg').convert("L")
IMG_SIZE = 512
resize_t = transforms.Resize((IMG_SIZE, IMG_SIZE), interpolation=PIL.Image.NEAREST)
imagen = resize_t(imagen)
plt.imshow(imagen)
```

```python
h, w = IMG_SIZE, IMG_SIZE
torch_image = torch.tensor(np.array(imagen))
torch_image = torch_image.view(1, h, w).reshape(1, 1, h, w)
print(torch_image.shape)
```

```python
from torchvision import transforms

def DHT(input):
    #input.type(torch.complex64)
    fft = torch.fft.fft2(input)
    dht = fft.real - fft.imag
    return dht
    

def torch_DiscreteHartleyTransform(input):
    
    dht = DHT(input)
    
    #print(dht)
    
    dht = torch.fft.fftshift(dht)
    #_dht = torch.fft.fftshift(torch.abs(dht))
    croped_dht = dht[:, :, IMG_SIZE//4:3*IMG_SIZE//4, IMG_SIZE//4:3*IMG_SIZE//4]
    croped_dht = torch.fft.ifftshift(croped_dht)
    #croped_dht = _spectral_crop(dht, IMG_SIZE//2, IMG_SIZE//2)
    
    dht = DHT(croped_dht)
    
    print(dht)
    
    return dht

dht = torch_DiscreteHartleyTransform(torch_image)
#dht = dht[0, 0, :, :].type(torch.float32)
#dht_pil = transforms.ToPILImage()(dht)
#dht_pil = transforms.ToPILImage()(dht)
#plt.imshow(dht_pil, cmap='gray')
print(dht.dtype)
plt.imshow(dht[0, 0, :, :].numpy().astype(np.float64), cmap='gray')
```

```python
def DHT_NP(input):
    fft = np.fft.fft2(input)
    dht = fft.real - fft.imag
    return dht

def np_DiscreteHartleyTransform(input):

    dht = DHT_NP(np.array(input))
    
    dht = np.fft.fftshift(dht)
    #dht = 20*np.log(np.abs(dht))
    croped_dht = dht[IMG_SIZE//4:3*IMG_SIZE//4, IMG_SIZE//4:3*IMG_SIZE//4]
    croped_dht = np.fft.ifftshift(croped_dht)
    
    dht = DHT_NP(croped_dht)

    return dht

dht = np_DiscreteHartleyTransform(imagen)
print(dht.dtype)
plt.imshow(dht, cmap='gray')
```
